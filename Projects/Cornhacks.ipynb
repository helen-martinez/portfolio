A Data Analysis of How Climate Change has Affected the Lives in the United States of America

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
```

#Cancer Data Set EDA

```python
cancer_etc = '/content/30-70cancerChdEtc.csv'

cancer = pd.read_csv(cancer_etc)

cancer = cancer.rename(columns={'First Tooltip': 'Percentage'})
cancer2 = cancer.loc[(cancer['Location'] == 'United States of America') & (cancer['Dim1'] == 'Both sexes')]

percentage_lst = []
for i in cancer2['Percentage']:
  percentage_lst.append(i)

period_lst = []
for i in cancer2['Period']:
  period_lst.append(i)

y_values = percentage_lst
x_values = period_lst

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Probability of dying between age 30-70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease in the U.S.')

# Create a line graph
plt.plot(x_values, y_values, marker='o', linestyle='-', color='blue', label='Line 1')

# Add a legend (optional)
plt.legend()

# Show the graph
plt.show()
```

# CO2 Emission rate

```python
co2_emission = '/content/CO2_emission.csv'

co2 = pd.read_csv(co2_emission)

co22 = co2.loc[(co2['Country Name'] == 'United States')]

# Assuming df is your DataFrame
# Filter only the columns that represent years
year_columns = co2.filter(regex=r'\b\d{4}\b').columns

# Replace 'your_country_name' with the desired country name
selected_country = co2[co2['Country Name'] == 'United States']

# Extract the values for the selected country
y_values = selected_country.iloc[:, 4:].values.flatten()
x_values = year_columns.astype(float)

# Create a line graph
plt.plot(x_values, y_values, marker='o', linestyle='-', color='blue', label='CO2 emissions')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('CO2 emissions (metric tons per capita)')
plt.title(f'CO2 Emissions Over Years for {selected_country.iloc[0]["Country Name"]}')
```

# Life Expectancy At Birth

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt

life_expectancy = '/content/lifeExpectancyAtBirth.csv'

leb = pd.read_csv(life_expectancy)

leb = leb.rename(columns={'First Tooltip': 'Age'})
leb2 = leb.loc[(leb['Location'] == 'United States of America') & (leb['Dim1'] == 'Both sexes')] #only US data

age_lst = []
for i in leb2['Age']:
  age_lst.append(i)

period_lst = []
for i in leb2['Period']:
  period_lst.append(i)

y_values = age_lst
x_values = period_lst

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Age')
plt.title('Life Expectancy in the U.S.')

# Create a line graph
plt.plot(x_values, y_values, marker='o', linestyle='-', color='blue', label='Line 1')

# Add a legend (optional)
plt.legend()

# Show the graph
plt.show()
```

# Maternal Mortality Rate

```python
maternal_mortality = '/content/maternalMortalityRatio.csv'

mm = pd.read_csv(maternal_mortality)

mm = mm.rename(columns={'First Tooltip': '# per 100K Pop.'})
mm2 = mm.loc[(mm['Location'] == 'United States of America')]

pop_lst = []
for i in mm2['# per 100K Pop.']:
  pop_lst.append(i)

period_lst = []
for i in mm2['Period']:
  period_lst.append(i)

y_values = pop_lst
x_values = period_lst

# Add labels and title
plt.xlabel('Year')
plt.ylabel('# per 100K Pop')
plt.title('Maternal Mortality Rate in the U.S.')

# Create a line graph
plt.plot(x_values, y_values, marker='o', linestyle='-', color='blue', label='Line 1')

# Add a legend (optional)
plt.legend()

# Show the graph
plt.show()
```

# Overall Takeaway: We found that as CO2 emissions have decreased, probability of death due to air pollution-related diseases decreased. Thus life expectancy increased. We also saw that maternal mortality rates decreased. There may be a correlation between these datasets. As our CO2 emissions decrease so have health-related issues decreased.
